## **BeautifulSoup**

### Import Module

```python
# Import necessary libraries

import requests #Sending and receiving HTTP requests
from bs4 import BeautifulSoup #Parsing and navigation of HTML and XML web pages
import pandas as pd #Used for data manipulation and analysis
```

### Get Response from URL

```python
# URL for the webpage to be scraped
url = 'https://www.petsecure.com.au/cover-choices/'

# Send a request to the webpage
response = requests.get(url)

# Print response
print(response)
```

### Parse & Print HTML with BeautifulSoup


```python
# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, 'html.parser')

print(soup.prettify()[:1000])
```

### Get Table Data from BeautifoulSoup

```python
# Get Div with Class table-2
assurance = soup.find('div', {'class': 'table-2'})
# tables = soup.find_all('table')
# cat_table=None

# for table in tables:
#     if "Cat" in table.text:
#         cat_table = table
#         break

# if cat_table:
#     print("Cat table found")
# else:
#     print("Cat table not found")



```

```python
# Get the table and print out
table = assurance.find('table')
print(table.prettify())

# if cat_table:
#     print("Cat table found")
# else:
#     print("Cat table not found")
```

### Extract Table Headers and Rows Data

```python
# Understanding How to get table headers

x = table.find_all('th')
for i, y in enumerate(x):
  print(i, y)
  print(i, "text:", y.text)
# print(table)
```

```python
# Extract the table headers
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())

print(headers)
```

```python
# Understanding How to get table headers

x = table.find_all('tr')
for i, y in enumerate(x):
  print(i, y.prettify())
  print(i, "text:", y.text)
# print(table)
```

```python
# Extract the table data
data = []
for tr in table.find_all('tr'):
    row_data = []
    for td in tr.find_all('td'):
        row_data.append(td.text.strip())
    if len(row_data) > 0:
        data.append(row_data)
```

```python
data
```

### Clean the Data (Optional)

```python
# List data to DataFrame

df = pd.DataFrame(data, columns = headers)
df
```

```python
# Check dataframe info

df.info()
```

```python
# Drop Column2 that contains nothing
df = df.drop(df[''],axis=1)
```

```python
# Drop '' in data
df.drop(df[df['Benefits']==''].index, inplace=True)
```
